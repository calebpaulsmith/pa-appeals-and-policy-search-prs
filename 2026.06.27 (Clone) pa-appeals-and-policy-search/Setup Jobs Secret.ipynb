```python
import requests

token = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()
host  = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiUrl().get()

resp = requests.post(
    f"{host}/api/2.0/secrets/scopes/create",
    headers={"Authorization": f"Bearer {token}"},
    json={"scope": "pa-appeals-and-policy-search"}
)
print(resp.status_code, resp.text)
# Expected: 200 {}  (or 400 RESOURCE_ALREADY_EXISTS if scope already exists — that's fine)
```

```python
YOUR_PAT = ""  # PAT cleared — already stored in Databricks secret

resp = requests.post(
    f"{host}/api/2.0/secrets/put",
    headers={"Authorization": f"Bearer {token}"},
    json={
        "scope": "pa-appeals-and-policy-search",
        "key": "jobs-token",
        "string_value": YOUR_PAT
    }
)
print(resp.status_code, resp.text)
# Expected: 200 {}
```

```python
# Confirm the key exists (value is always redacted)
keys = [s.key for s in dbutils.secrets.list("pa-appeals-and-policy-search")]
print("Keys in scope:", keys)
# Expected: ['jobs-token']
```
